```python
import skimage.io as io

# Show the figures / plots inside the notebook
%matplotlib inline
from skimage.color import rgb2gray,rgb2hsv
import matplotlib.pyplot as plt
import numpy as np
from skimage.util import random_noise
import numpy as np

from skimage.exposure import histogram
from matplotlib.pyplot import bar
```

```python

def show_images(images,titles=None):
    #This function is used to show image(s) with titles by sending an array of images and an array of associated titles.
    # images[0] will be drawn with the title titles[0] if exists
    # You aren't required to understand this function, use it as-is.
    n_ims = len(images)
    if titles is None: titles = ['(%d)' % i for i in range(1,n_ims + 1)]
    fig = plt.figure()
    n = 1
    for image,title in zip(images,titles):
        a = fig.add_subplot(1,n_ims,n)
        if image.ndim == 2: 
            plt.gray()
        plt.imshow(image)
        a.set_title(title)
        n += 1
    fig.set_size_inches(np.array(fig.get_size_inches()) * n_ims)
    plt.show()
```

```python
"""
@TODO
Requirement 2 (Noise):
"""

def Adding_S_P_Noise(image,mode,amount) -> np.ndarray:
    return random_noise(image, mode=mode, amount=amount)


# Read the image.

img = io.imread("./Abdelrahman_Photo.png")
img = img[:,:,:3]

# Convert it to greyscale to the predefined image (img) and showing it

gray_img = rgb2gray(img)

images = [img,gray_img]
titles = ["Original Image","Gray-scaled Image"]
show_images(images, titles)


# Apply salt & pepper noise to the predefined image (img)

first_pixeled_img = Adding_S_P_Noise(img,"s&p",0.05)
second_pixeled_img = Adding_S_P_Noise(img,"s&p",0.5)
third_pixeled_img = Adding_S_P_Noise(img,"s&p",0.9)

images = [first_pixeled_img,second_pixeled_img,third_pixeled_img]
titles=["First Piexeled Image","Second Piexeled Image","Third Piexeled Image"]
show_images(images,titles)



'''
    - The "random_noise" function simply it is  function that introduces the property of adding a noise to the passed image
    with different modes in order to ensure the error resilience and robustness of algorithms and models.

    - The image that won't be affected by noise too much is the first image "first_pixeled_image", as the amount(fraction) of pixels endured
    from noise are only 5% and that is the smallest factor among the other values. 
'''
```

```python
'''
@TODO
Requirement 3 (Histogram):
'''
def showHist(images, titles):
    plt.figure(figsize=(15, 4))
    plt.style.use("dark_background")

    plt.subplot(131)
    plt.title(titles[0])
    plt.xlabel("Pixel Intensity")
    plt.ylabel("Frequency")
    plt.fill_between(images[0][1], images[0][0],alpha=0.5)

    if len(images)>1:
        plt.subplot(132)
        plt.title(titles[1])
        plt.xlabel("Pixel Intensity")
        plt.ylabel("Frequency")
        plt.fill_between(images[1][1], images[1][0], alpha=0.5)

        plt.subplot(133)
        plt.title(titles[2])
        plt.xlabel("Pixel Intensity")
        plt.ylabel("Frequency")
        plt.fill_between(images[2][1], images[2][0], alpha=0.5)


def generateHistograms (source_images):
    for i in range(len(source_images)):
        images = []
        for unit_bin in [256, 64, 8]:
            hist, bins = histogram(rgb2gray((source_images[i])[:,:,:3]), nbins=unit_bin)
            images.append((hist, bins))
        if i == 0:
            titles = ["First Histogram-256","First Histogram-64","First Histogram-8"]
        elif i == 1:
            titles = ["Second Histogram-256", "Second Histogram-64", "Second Histogram-8"]
        else:
            titles = ["Third Histogram-256","Third Histogram-64","Third Histogram-8"]
        showHist(images,titles)


img_1 = io.imread("./histogram/ex1.jpg")
img_2 = io.imread("./histogram/ex2.png")
img_3 = io.imread("./histogram/ex3.png")

# The nbins parameter means that each intensity will be divided to nbins-number such that each intensity value will be counted up for the pixels in range of ( i -> i + (nbins / 256) - 1 )

source_images = [img_1,img_2,img_3]
images = generateHistograms(source_images)
```
